## 1. Sobre el paper

El artículo "Ensemble Boosted Trees with Synthetic Features Generation in Application to Bankruptcy Prediction" propone un nuevo enfoque para la predicción de la bancarrota empresarial mediante técnicas de aprendizaje automático. El objetivo principal de la investigación es desarrollar un modelo capaz de identificar con mayor precisión las empresas con riesgo de quiebra utilizando indicadores financieros, superando las limitaciones de los métodos estadísticos tradicionales y de otros algoritmos de clasificación ampliamente utilizados.

La motivación del estudio surge de dos problemas comunes en este tipo de aplicaciones. En primer lugar, la dificultad de combinar adecuadamente los numerosos indicadores financieros utilizados para describir la situación económica de una empresa. En segundo lugar, el fuerte desbalance existente en este tipo de conjuntos de datos, donde las empresas que no quiebran representan la gran mayoría de las observaciones, lo que puede afectar negativamente el entrenamiento de los modelos predictivos.

Para abordar estos problemas, los autores proponen un modelo basado en Extreme Gradient Boosting (XGBoost), un algoritmo de aprendizaje por ensamble que construye múltiples árboles de decisión de manera secuencial. Como aporte adicional, el artículo introduce un mecanismo de generación de atributos sintéticos, mediante el cual se crean nuevas variables a partir de combinaciones aritméticas de los indicadores financieros originales. Estas nuevas características buscan representar relaciones más complejas entre las variables y mejorar la capacidad predictiva del modelo.

Los experimentos se realizaron utilizando información financiera de empresas manufactureras polacas obtenida de la base de datos EMIS. El conjunto de datos incluye 64 indicadores financieros y cinco escenarios de predicción, correspondientes a la estimación de la bancarrota entre uno y cinco años antes de que ocurra el evento.

Como referencia, los autores compararon su propuesta con diversos modelos de clasificación, entre ellos Regresión Logística, Máquinas de Vectores de Soporte (SVM), Redes Neuronales Multicapa (MLP), Random Forest, AdaBoost y árboles de decisión. La evaluación se realizó utilizando validación cruzada de 10 particiones y el área bajo la curva ROC (AUC), una métrica apropiada para problemas de clasificación con datos desbalanceados.

Los resultados muestran que el modelo basado en XGBoost supera consistentemente a los demás métodos evaluados y que la incorporación de atributos sintéticos mejora aún más su desempeño. Los autores concluyen que esta estrategia permite capturar relaciones no lineales entre los indicadores financieros y constituye una alternativa eficaz para la predicción temprana de la bancarrota empresarial.

El procedimiento consiste en generar múltiples expresiones simbólicas mediante operadores aritméticos como suma, resta, multiplicación y división. Posteriormente, aquellas expresiones con mayor capacidad predictiva son seleccionadas como nuevos atributos y se incorporan al conjunto de entrenamiento. De esta forma, el modelo dispone de una representación más rica de los datos, siguiendo la misma idea propuesta en el artículo aunque utilizando una implementación diferente.

**Extracción de diversidad (importancia de variables)**

Después de generar los atributos sintéticos, el notebook analiza las expresiones obtenidas con el objetivo de identificar cuáles variables financieras aparecen con mayor frecuencia en las fórmulas generadas por la Programación Genética.

Este análisis permite estimar la relevancia de cada indicador financiero dentro del proceso de construcción de los nuevos atributos. En términos prácticos, una mayor frecuencia de aparición sugiere que determinada variable aporta información útil para diferenciar entre empresas con riesgo de bancarrota y empresas financieramente estables.

Aunque este procedimiento no reproduce exactamente el mecanismo de importancia de variables utilizado en el artículo original, persigue el mismo objetivo: comprender qué indicadores financieros tienen mayor influencia en la capacidad predictiva del modelo. En el paper, esta importancia se calcula a partir de la frecuencia con la que las variables aparecen en los árboles del ensamble y en los atributos sintéticos generados.

```python
# ==============================================================================
# PARSER CORREGIDO
# ==============================================================================
def prefix_to_infix(expr_str):
    if str(expr_str) == "None" or "Invalid" in str(expr_str):
        return "Invalid Program"

    expr_str = expr_str.replace(" ", "")
    # CORRECCIÓN: Se agregó '\)' al regex para no omitir los cierres en expresiones anidadas
    tokens = re.findall(r'[a-zA-Z0-9_.]+|[\(\,\)]', expr_str)

    def parse(index):
        if index >= len(tokens):
            return "", index
        token = tokens[index]
        if token in ('add', 'sub', 'mul', 'div'):
            op_map = {'add': '+', 'sub': '-', 'mul': '*', 'div': '/'}
            op = op_map[token]
            index += 2  # Salta función y '('
            arg1, index = parse(index)
            index += 1  # Salta ','
            arg2, index = parse(index)
            index += 1  # Salta ')'
            return f"({arg1} {op} {arg2})", index
        else:
            match = re.match(r'[xX](\d+)', token)
            if match:
                var_idx = int(match.group(1)) + 1
                token = f"X{var_idx}"
            return token, index + 1

    try:
        result, _ = parse(0)
        return result
    except:
        return expr_str
```

En esta celda se configura y entrena el modelo XGBoost utilizando validación cruzada de 10 particiones. Para cada división del conjunto de datos se ajusta el modelo con los datos de entrenamiento y posteriormente se evalúa sobre el conjunto de prueba mediante la métrica AUC. Este procedimiento permite estimar de manera más confiable el desempeño del algoritmo y reducir el riesgo de sobreajuste.

Igualmente, en esta parte se aplica el algoritmo SymbolicTransformer de la biblioteca gplearn para generar automáticamente atributos sintéticos mediante Programación Genética. El algoritmo combina los indicadores financieros originales utilizando operadores matemáticos y selecciona aquellas expresiones que presentan mayor capacidad para discriminar entre las clases. Los nuevos atributos obtenidos se incorporan posteriormente al conjunto de datos utilizado por el modelo de clasificación.

```python
# ==============================================================================
# ENTRENAMIENTO Y EXTRACCIÓN DE DIVERSIDAD
# ==============================================================================
def evaluate_year(file_path, year_label):
    print(f"\n{'='*40}")
    print(f"PROCESANDO DATOS DEL: {year_label}")
    print(f"{'='*40}")

    data, meta = arff.loadarff(file_path)
    df = pd.DataFrame(data)
    df['class'] = df['class'].astype(int)

    X = df.drop('class', axis=1).values
    y = df['class'].values

    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
    auc_scores = []

    fold_programs = []
    fold_xgb_model = None

    for train_idx, test_idx in tqdm(skf.split(X, y), total=10, desc=f"Folds de {year_label}"):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        imputer = SimpleImputer(strategy='mean')
        X_train_imp = imputer.fit_transform(X_train)
        X_test_imp = imputer.transform(X_test)

        # Configuración para fórmulas más compactas y parecidas al paper
        gp = SymbolicTransformer(
            generations=15,
            population_size=1000,
            hall_of_fame=200,
            n_components=100,
            function_set=['add', 'sub', 'mul', 'div'],
            init_depth=(1, 3), # Previene over-fitting matemático
            const_range=None,  # Obliga a operar solo con las variables originales
            parsimony_coefficient=0.005,
            verbose=0,
            random_state=42
        )
        gp.fit(X_train_imp, y_train)

        # CORRECCIÓN: Filtrar para obtener exactamente 60 programas ÚNICOS
        unique_programs = []
        unique_strs = set()

        for p in gp:
            p_str = str(p)
            if p is not None and p_str not in unique_strs:
                unique_strs.add(p_str)
                unique_programs.append(p)
            if len(unique_programs) == 60:
                break

        # Fallback si _best_programs no tiene suficientes únicos
        if len(unique_programs) < 60:
            for gen in gp._programs:
                if isinstance(gen, list):
                    for p in gen:
                        p_str = str(p)
                        if p is not None and p_str not in unique_strs:
                            unique_strs.add(p_str)
                            unique_programs.append(p)
                        if len(unique_programs) == 60: break
                if len(unique_programs) == 60: break

        # Ejecutar manualmente las transformaciones sobre los programas extraídos
        syn_train = np.column_stack([p.execute(X_train_imp) for p in unique_programs])
        syn_test = np.column_stack([p.execute(X_test_imp) for p in unique_programs])

        X_train_syn = np.hstack((X_train_imp, syn_train))
        X_test_syn = np.hstack((X_test_imp, syn_test))

        ratio = float(np.sum(y_train == 0)) / np.sum(y_train == 1)
        xgb_model = XGBClassifier(
            n_estimators=100, max_depth=5, learning_rate=0.1,
            scale_pos_weight=ratio, random_state=42, eval_metric='auc', n_jobs=-1
        )
        xgb_model.fit(X_train_syn, y_train)

        y_pred_proba = xgb_model.predict_proba(X_test_syn)[:, 1]
        auc = roc_auc_score(y_test, y_pred_proba)
        auc_scores.append(auc)

        fold_programs = unique_programs
        fold_xgb_model = xgb_model

    mean_auc = np.mean(auc_scores)
    std_auc = np.std(auc_scores)

    print(f"\nResultados finales para {year_label}:")
    print(f"ROC-AUC Mean (MN) : {mean_auc:.4f}")
    print(f"ROC-AUC Std (STD) : {std_auc:.4f}\n")

    return mean_auc, std_auc, fold_programs, fold_xgb_model
```

Finalmente, se analizan las expresiones simbólicas generadas para identificar la frecuencia con la que aparece cada variable financiera. Este análisis permite interpretar parcialmente el proceso de generación de atributos sintéticos e identificar cuáles indicadores originales contribuyen con mayor frecuencia a la construcción de nuevas características predictivas.

## 4. Modelos implementados
**Modelo principal: Extreme Gradient Boosting (XGBoost)**

El modelo principal implementado en esta réplica es Extreme Gradient Boosting (XGBoost), ya que constituye la metodología central propuesta por los autores del artículo. XGBoost es un algoritmo de aprendizaje por ensamble basado en árboles de decisión que construye secuencialmente múltiples modelos débiles, donde cada nuevo árbol intenta corregir los errores cometidos por los anteriores. Esta estrategia permite obtener modelos altamente precisos y con buena capacidad para capturar relaciones no lineales entre las variables financieras.

Una de las principales ventajas de XGBoost es la incorporación de mecanismos de regularización que reducen el sobreajuste y mejoran la capacidad de generalización. Además, el algoritmo permite optimizar directamente métricas apropiadas para conjuntos de datos desbalanceados, como el Área Bajo la Curva ROC (AUC), característica especialmente importante en problemas de predicción de bancarrota.

En esta implementación el modelo se evalúa mediante validación cruzada de diez particiones, calculando el AUC promedio como medida del desempeño predictivo.

**Modelo adicional: LightGBM**

Como parte de los requisitos del proyecto se implementa un modelo adicional denominado Light Gradient Boosting Machine (LightGBM).

LightGBM también pertenece a la familia de algoritmos de Gradient Boosting, pero incorpora una estrategia de crecimiento de árboles basada en hojas (leaf-wise growth), lo que generalmente permite entrenar modelos más rápidos y eficientes cuando se trabaja con grandes volúmenes de datos.

El objetivo de incorporar este algoritmo es comparar su capacidad predictiva con la obtenida por XGBoost utilizando exactamente el mismo conjunto de datos y el mismo procedimiento de validación. De esta manera es posible analizar si un algoritmo moderno de boosting ofrece ventajas adicionales sobre el método utilizado en el artículo original.

**Justificación de los modelos seleccionados**

La elección de XGBoost responde directamente a la metodología propuesta por los autores del artículo, quienes demostraron que este algoritmo obtuvo los mejores resultados frente a otros modelos clásicos como Regresión Logística, Random Forest, AdaBoost y Máquinas de Vectores de Soporte.

Por otra parte, LightGBM fue seleccionado como modelo alternativo debido a que comparte principios similares de aprendizaje por ensamble, pero utiliza una estrategia de optimización diferente. Esta comparación permite evaluar si las mejoras introducidas por LightGBM producen un desempeño comparable o superior sobre el mismo problema de clasificación.

```python
# ==============================================================================
# EJECUCIÓN PRINCIPAL
# ==============================================================================
archivos = [
    ('polish_data/1year.arff', '1st Year'),
    ('polish_data/2year.arff', '2nd Year'),
    ('polish_data/3year.arff', '3rd Year'),
    ('polish_data/4year.arff', '4th Year'),
    ('polish_data/5year.arff', '5th Year')
]

resultados_finales = {}
final_programs = None
final_xgb_model = None

for ruta, nombre_ano in archivos:
    if os.path.exists(ruta):
        mn_auc, std_auc, current_programs, current_xgb_model = evaluate_year(ruta, nombre_ano)
        resultados_finales[nombre_ano] = {'Mean AUC': mn_auc, 'Std AUC': std_auc}
        final_programs = current_programs
        final_xgb_model = current_xgb_model
    else:
        print(f"No se encontró el archivo: {ruta}")

print("\n" + "="*50)
print("RESUMEN DE RESULTADOS (10-FOLD CV)")
print("="*50)
resumen_df = pd.DataFrame(resultados_finales).T
print(resumen_df)
```

**Interpretación de los resultados del entrenamiento**

## 6. Análisis de interpretabilidad
**Generación e interpretación de atributos sintéticos**

Una de las ventajas de utilizar Programación Genética es que los atributos sintéticos generados pueden expresarse como fórmulas matemáticas interpretables. A diferencia de otros métodos de ingeniería de características, estas expresiones permiten identificar qué indicadores financieros participan en la construcción de las nuevas variables y cómo se combinan entre sí.

En esta etapa del experimento se analizan las características sintéticas con mayor importancia para el modelo. Para ello, se identifican las variables originales que intervienen en cada expresión y se traduce cada fórmula a una representación más comprensible desde el punto de vista financiero.

Este análisis facilita la interpretación del modelo y permite relacionar los resultados obtenidos con indicadores financieros relevantes para la predicción de la bancarrota, manteniendo el objetivo del artículo original de combinar un alto desempeño predictivo con una mejor comprensión de las variables más influyentes.

```python
# ==============================================================================
# ANÁLISIS DE INTERPRETABILIDAD CORREGIDO
# ==============================================================================
def analyze_interpretability(unique_programs, xgb_classifier):
    print("\n" + "="*80)
    print(" EXTRACCIÓN DE FÓRMULAS SINTÉTICAS E INFERENCIA ECONÓMICA (CORREGIDO) ")
    print("="*80)

    num_synthetic_features = len(unique_programs)
    synthetic_formulas_infix = [prefix_to_infix(str(p)) for p in unique_programs]

    num_original = 64
    feature_labels = [f"X{i}" for i in range(1, num_original + 1)]
    synthetic_labels = [f"S{i}" for i in range(1, num_synthetic_features + 1)]
    all_features = feature_labels + synthetic_labels

    importances = xgb_classifier.feature_importances_

    ranking_df = pd.DataFrame({
        'Feature': all_features,
        'Importance': importances,
        'Type': ['Original' if i < num_original else 'Synthetic' for i in range(len(all_features))]
    })

    ranking_df = ranking_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)

    print("\n--- TOP 10 VARIABLES MÁS IMPORTANTES PARA LA PREDICCIÓN ---")
    print(ranking_df.head(10).to_string(index=False))
    print("\n" + "-"*80)

    print("\n--- DECODIFICACIÓN DE LAS VARIABLES SINTÉTICAS MÁS INFLUYENTES ---")
    synthetic_rankings = ranking_df[ranking_df['Type'] == 'Synthetic'].head(5)

    if len(synthetic_rankings) == 0:
        print("No hay variables sintéticas dentro del ranking destacado de XGBoost.")
        return

    for idx, row in synthetic_rankings.iterrows():
        feat_name = row['Feature']
        importance = row['Importance']
        s_idx = int(feat_name[1:]) - 1

        formula_math = synthetic_formulas_infix[s_idx]

        print(f"\n⚡ VARIABLE SINTÉTICA: {feat_name} (Importancia relativa: {importance:.4%})")
        print(f"   Fórmula algebraica: {formula_math}")
        print("   Componentes financieros involucrados:")

        components = sorted(list(set(re.findall(r'X\d+', formula_math))), key=lambda x: int(x[1:]))
        for comp in components:
            comp_idx = int(comp[1:])
            description = FINANCIAL_RATIOS_DICT.get(comp_idx, "Descripción no disponible")
            print(f"      • {comp}: {description}")

    print("\n" + "="*80)

# Lanzar el análisis usando el modelo y los programas de la última iteración
if final_programs and final_xgb_model:
    analyze_interpretability(final_programs, final_xgb_model)
```

**Interpretación de los resultados**

El análisis de interpretabilidad muestra que las variables sintéticas generadas mediante Programación Genética tuvieron una contribución importante en el proceso de clasificación. En particular, la característica S53 presentó la mayor importancia relativa (16.61 %), superando incluso a varias variables financieras originales. Esto indica que la combinación de múltiples indicadores puede capturar relaciones que no son evidentes cuando cada variable se analiza de manera independiente.

Las fórmulas generadas combinan indicadores relacionados con liquidez, rentabilidad, estructura de capital, eficiencia operativa y tamaño de los activos, lo que coincide con los factores financieros que tradicionalmente se consideran relevantes para evaluar el riesgo de bancarrota.

Por ejemplo, la variable sintética S53 integra indicadores de ventas, participación del patrimonio sobre los activos y rentabilidad sobre activos, sugiriendo que la interacción entre crecimiento comercial, estructura financiera y rentabilidad aporta información valiosa para diferenciar empresas con mayor probabilidad de quiebra.

Asimismo, otras variables sintéticas relevantes incorporan razones financieras relacionadas con el nivel de activos, el capital de trabajo, la rotación de inventarios y las obligaciones de corto plazo. Estos resultados muestran que el algoritmo de Programación Genética fue capaz de construir nuevas características que combinan distintos aspectos de la situación financiera de las empresas, enriqueciendo la representación de los datos utilizada por el modelo predictivo.

En conjunto, este análisis demuestra que la generación automática de atributos no solo mejora la capacidad predictiva del modelo, sino que también proporciona una herramienta útil para interpretar qué relaciones financieras son más relevantes en la identificación temprana del riesgo de bancarrota.

## 7. Comparación entre XGBoost y LightGBM

Continuando con el análisis, se amplía el experimento mediante la incorporación de LightGBM como modelo adicional. El objetivo es comparar su desempeño con XGBoost, que corresponde al algoritmo principal utilizado en el artículo original.

Para asegurar una comparación justa, ambos modelos se entrenan con:

Las mismas particiones de validación cruzada estratificada.
Los mismos datos imputados.
Las mismas variables financieras originales.
Las mismas características sintéticas generadas mediante Programación Genética.
El mismo tratamiento del desbalance de clases.
La misma métrica de evaluación: ROC-AUC.

De esta manera, las diferencias observadas en el rendimiento pueden atribuirse principalmente al comportamiento de cada algoritmo de boosting y no a cambios en los datos o en el procedimiento de evaluación.

**Procedimiento experimental**

Para cada uno de los cinco horizontes temporales, el código realiza una validación cruzada estratificada de 10 particiones. Dentro de cada partición se ejecutan las siguientes etapas:

1. Se dividen los datos en entrenamiento y prueba.
2. Los valores faltantes se imputan utilizando la media calculada únicamente sobre el conjunto de entrenamiento.
3. Se entrena un *SymbolicTransformer* para generar características sintéticas.
4. Se seleccionan hasta 60 expresiones simbólicas únicas.
5. Las nuevas características se agregan a las variables financieras originales.
6. Se calcula la relación entre clases para compensar el desbalance.
7. Se entrenan XGBoost y LightGBM sobre exactamente la misma matriz de características.
8. Se calcula el ROC-AUC de ambos modelos sobre el conjunto de prueba.

Al finalizar las diez particiones, se reportan el AUC promedio y su desviación estándar para cada algoritmo.

```python
from lightgbm import LGBMClassifier

def evaluate_year(file_path, year_label):
    print(f"\n{'='*40}")
    print(f"PROCESANDO DATOS DEL: {year_label}")
    print(f"{'='*40}")

    data, meta = arff.loadarff(file_path)
    df = pd.DataFrame(data)
    df['class'] = df['class'].astype(int)

    X = df.drop('class', axis=1).values
    y = df['class'].values

    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    # Listas para guardar los scores de ambos modelos
    auc_scores_xgb = []
    auc_scores_lgbm = []

    fold_programs = []
    fold_xgb_model = None

    for train_idx, test_idx in tqdm(skf.split(X, y), total=10, desc=f"Folds de {year_label}"):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        imputer = SimpleImputer(strategy='mean')
        X_train_imp = imputer.fit_transform(X_train)
        X_test_imp = imputer.transform(X_test)

        # 1. Transformador Genético
        gp = SymbolicTransformer(
            generations=15,
            population_size=1000,
            hall_of_fame=200,
            n_components=100,
            function_set=['add', 'sub', 'mul', 'div'],
            init_depth=(1, 3),
            const_range=None,
            parsimony_coefficient=0.005,
            verbose=0,
            random_state=42
        )
        gp.fit(X_train_imp, y_train)

        # 2. Extracción de Programas Únicos
        unique_programs = []
        unique_strs = set()

        for p in gp:
            p_str = str(p)
            if p is not None and p_str not in unique_strs:
                unique_strs.add(p_str)
                unique_programs.append(p)
            if len(unique_programs) == 60:
                break

        if len(unique_programs) < 60:
            for gen in gp._programs:
                if isinstance(gen, list):
                    for p in gen:
                        p_str = str(p)
                        if p is not None and p_str not in unique_strs:
                            unique_strs.add(p_str)
                            unique_programs.append(p)
                        if len(unique_programs) == 60: break
                if len(unique_programs) == 60: break

        # Creación de variables sintéticas
        syn_train = np.column_stack([p.execute(X_train_imp) for p in unique_programs])
        syn_test = np.column_stack([p.execute(X_test_imp) for p in unique_programs])

        X_train_syn = np.hstack((X_train_imp, syn_train))
        X_test_syn = np.hstack((X_test_imp, syn_test))

        ratio = float(np.sum(y_train == 0)) / np.sum(y_train == 1)

        # ---------------------------------------------------------
        # 3. Entrenamiento XGBoost (Modelo del Paper)
        # ---------------------------------------------------------
        xgb_model = XGBClassifier(
            n_estimators=100, max_depth=5, learning_rate=0.1,
            scale_pos_weight=ratio, random_state=42, eval_metric='auc', n_jobs=-1
        )
        xgb_model.fit(X_train_syn, y_train)

        y_pred_proba_xgb = xgb_model.predict_proba(X_test_syn)[:, 1]
        auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)
        auc_scores_xgb.append(auc_xgb)

        # ---------------------------------------------------------
        # 4. Entrenamiento LightGBM (Modelo Adicional Propuesto)
        # ---------------------------------------------------------
        lgbm_model = LGBMClassifier(
            n_estimators=100, max_depth=5, learning_rate=0.1,
            scale_pos_weight=ratio, random_state=42, n_jobs=-1,
            verbose=-1 # Silencia las advertencias de LightGBM
        )
        lgbm_model.fit(X_train_syn, y_train)

        y_pred_proba_lgbm = lgbm_model.predict_proba(X_test_syn)[:, 1]
        auc_lgbm = roc_auc_score(y_test, y_pred_proba_lgbm)
        auc_scores_lgbm.append(auc_lgbm)

        # Guardar para análisis de interpretabilidad (usamos XGBoost como principal)
        fold_programs = unique_programs
        fold_xgb_model = xgb_model

    # Calcular promedios y desviaciones
    mean_auc_xgb = np.mean(auc_scores_xgb)
    std_auc_xgb = np.std(auc_scores_xgb)

    mean_auc_lgbm = np.mean(auc_scores_lgbm)
    std_auc_lgbm = np.std(auc_scores_lgbm)

    print(f"\nResultados finales para {year_label}:")
    print(f"XGBoost  -> ROC-AUC Mean: {mean_auc_xgb:.4f} | Std: {std_auc_xgb:.4f}")
    print(f"LightGBM -> ROC-AUC Mean: {mean_auc_lgbm:.4f} | Std: {std_auc_lgbm:.4f}\n")

    return mean_auc_xgb, std_auc_xgb, mean_auc_lgbm, std_auc_lgbm, fold_programs, fold_xgb_model
```

**Interpretación metodológica**

La función definida en esta celda permite realizar una comparación controlada entre XGBoost y LightGBM. Ambos modelos reciben exactamente la misma información en cada partición, incluyendo los atributos sintéticos generados exclusivamente a partir de los datos de entrenamiento.

Este último aspecto es importante, ya que evita utilizar información del conjunto de prueba durante la imputación o la generación de características. Por tanto, el procedimiento reduce el riesgo de fuga de información y proporciona una estimación más realista de la capacidad de generalización de los modelos.

La selección de programas únicos también evita incorporar repetidamente la misma fórmula sintética. Esto aumenta la diversidad del conjunto de características y reduce la redundancia entre las variables generadas.

Por otra parte, el parámetro scale_pos_weight se calcula en cada partición a partir de la relación entre empresas no quebradas y empresas quebradas. Este ajuste permite que los modelos presten mayor atención a la clase minoritaria, lo cual es relevante debido al fuerte desbalance presente en los datos.

Finalmente, XGBoost se conserva como modelo principal para el análisis de interpretabilidad, mientras que LightGBM se utiliza como modelo adicional de comparación. Esta decisión mantiene la relación con la metodología del artículo y, al mismo tiempo, satisface el objetivo de evaluar una alternativa moderna de Gradient Boosting.

## 8. Ejecución principal y comparación de modelos

En esta etapa se ejecuta el experimento completo para los cinco conjuntos de datos del Polish Companies Bankruptcy Dataset, correspondientes a los horizontes de predicción de uno a cinco años antes de la posible bancarrota.

Para cada conjunto de datos se aplica el mismo procedimiento experimental:

1. Se realiza una validación cruzada estratificada de 10 particiones.
2. Se imputan los valores faltantes utilizando la media del conjunto de entrenamiento.
3. Se generan atributos sintéticos mediante Programación Genética con gplearn.
4. Se entrenan los modelos XGBoost y LightGBM utilizando exactamente las mismas características y particiones.
5. Se evalúa el desempeño mediante el Área Bajo la Curva ROC (AUC).
6. Finalmente, se calcula el AUC promedio y su desviación estándar para cada algoritmo, construyendo una tabla comparativa que resume los resultados obtenidos.

El propósito de esta etapa es determinar si el modelo propuesto en el artículo (XGBoost) mantiene un desempeño superior cuando se compara con un modelo alternativo de Gradient Boosting bajo las mismas condiciones experimentales.

```python
# ==============================================================================
# EJECUCIÓN PRINCIPAL Y TABLA COMPARATIVA
# ==============================================================================
archivos = [
    ('polish_data/1year.arff', '1st Year'),
    ('polish_data/2year.arff', '2nd Year'),
    ('polish_data/3year.arff', '3rd Year'),
    ('polish_data/4year.arff', '4th Year'),
    ('polish_data/5year.arff', '5th Year')
]

resultados_finales = {}
final_programs = None
final_xgb_model = None

for ruta, nombre_ano in archivos:
    if os.path.exists(ruta):
        mn_xgb, std_xgb, mn_lgbm, std_lgbm, current_programs, current_xgb_model = evaluate_year(ruta, nombre_ano)

        # Guardamos ambos resultados en el diccionario para la tabla final
        resultados_finales[nombre_ano] = {
            'XGBoost (Paper) - Mean AUC': mn_xgb,
            'XGBoost (Paper) - Std AUC': std_xgb,
            'LightGBM (Propuesto) - Mean AUC': mn_lgbm,
            'LightGBM (Propuesto) - Std AUC': std_lgbm
        }

        final_programs = current_programs
        final_xgb_model = current_xgb_model
    else:
        print(f"No se encontró el archivo: {ruta}")

print("\n" + "="*80)
print(" TABLA COMPARATIVA DE MODELOS (10-FOLD CV) ".center(80))
print("="*80)
resumen_df = pd.DataFrame(resultados_finales).T
# Formatear el DataFrame para que sea más legible en el output
print(resumen_df.round(4).to_string())
print("="*80)
```

En términos generales, XGBoost obtuvo el mejor desempeño en la mayoría de los escenarios evaluados, alcanzando los mayores valores promedio de AUC en los conjuntos correspondientes al segundo, tercer, cuarto y quinto año de predicción.
